## Exercici 4

Has d'obtenir, per a cada persona, el seu nom i el nombre màxim de rols diferents que ha tingut en una mateixa pel·lícula. 
Posteriorment, mostra únicament aquelles persones que hagin assumit més d'un rol en una mateixa pel·lícula.

```python
delays_query = """ 
               SELECT movie_title as Título, person_name as Nombre, 
	count(role_name) as roles 
FROM tb_movie_person mp 
INNER JOIN tb_movie m 
	ON m.movie_id = mp.movie_id 
INNER JOIN tb_person p 
	ON p.person_id = mp.person_id 
INNER JOIN tb_role r 
	ON r.role_id = mp.role_id 
GROUP BY movie_title, person_name 
	HAVING roles > 1 ORDER BY roles DESC;    

               """
query3 = pd.read_sql(delays_query, con=mysql_engine)
query3
```
